```python
# import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# read csv file

df = pd.read_csv('Entertainer_Data_analysis.csv')
```

```python
# check for missing values

df.isnull().sum()
```

```python
# fill missing values with 0

df = df.fillna(0)
```

```python
# convert 'Year of Death' column to datetime format
df['Year of Death'] = pd.to_datetime(df['Year of Death'], errors='coerce', format='%Y')

df
```

```python
# create new column 'Age at Death'
df['Age at Death'] = df['Year of Death'].dt.year - df['Birth Year']

df
```

```python
# create new column 'Years since Last Work'

df['Years since Last Work'] = 2021 - df['Last work']

df
```

```python
# create new column 'Grammy Winner'

df['Grammy Winner'] = np.where(df['Grammy'] == 'Yes', 1, 0)

df
```

```python
# create bar plot for 'Breakthrough' column

sns.countplot(x='Breakthrough', data=df)
plt.title('Breakthrough in Career')
plt.xlabel('Breakthrough')
plt.ylabel('Count')
plt.show()
```

```python
# create scatter plot for 'Age at Death' and 'Years since Last Work' columns

sns.scatterplot(x='Age at Death', y='Years since Last Work', data=df)
plt.title('Age at Death vs Years since Last Work')
plt.xlabel('Age at Death')
plt.ylabel('Years since Last Work')
plt.show()
```

```python
# create bar plot for 'Grammy Winner' column

sns.countplot(x='Grammy Winner', data=df)
plt.title('Grammy Winners')
plt.xlabel('Grammy Winner')
plt.ylabel('Count')
plt.show()
```

```python
# create box plot for 'Age at Death' column

sns.boxplot(x='Age at Death', data=df)
plt.title('Age at Death Distribution')
plt.xlabel('Age at Death')
plt.show()
```

```python
# create box plot for 'Years since Last Work' column

sns.boxplot(x='Years since Last Work', data=df)
plt.title('Years since Last Work Distribution')
plt.xlabel('Years since Last Work')
plt.show()
```

```python
# create bar plot for 'Gender' column

sns.countplot(x='Gender', data=df)
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
# create bar plot for 'Breakthrough Name' column

plt.figure(figsize=(10,6))
sns.countplot(x='Breakthrough Name', data=df)
plt.title('Breakthrough Names')
plt.xlabel('Breakthrough Name')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

```python
# create heatmap to check correlation between numerical columns

sns.heatmap(df[['Age at Death', 'Years since Last Work', 'Grammy Winner']].corr(), annot=True)
plt.title('Correlation Heatmap')
plt.show()
```
